```python
import pandas as pd 
```

```python
import numpy as np
```

```python
estados = pd.read_csv('brasil_estados.csv')
```

```python
estados.groupby('regiao_nome')['populacao'].sum()
```

```python
round(estados['area'].describe(), 2)
```

```python
estados['area'].plot()
```

```python
import seaborn as sns
```

```python
sns.boxplot(
    data=estados, 
    x='regiao_nome', 
    y='area'
    )
```

```python
out_max = estados['area'].mean() + (3 * estados['area'].std())
```

```python
print(out_max)
```

```python
out_max_mad = estados['area'].quantile(0.75) + ((estados['area'].quantile(0.75) - estados['area'].quantile(0.25)) * 1.5)
```

```python
print(out_max_mad)
```

```python
limite_inferior = estados['area'].mean() - (1.96 * estados['area'].std() / np.sqrt(len(estados['area'])))
```

```python
print(limite_inferior)
```

```python
limite_superior = estados['area'].mean() + (1.96 * estados['area'].std() / np.sqrt(len(estados['area'])))
```

```python
print(limite_superior)
```
